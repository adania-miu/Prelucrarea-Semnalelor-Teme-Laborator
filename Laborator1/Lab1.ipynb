### 1. Fie semnalele continue 
x(t) = cos(520 * pi * t + pi / 3)

y(t) = cos(280 * pi * t - pi / 3)

z(t) = cos(120 * pi * t + pi / 3)

### a) In Python, simulati axa reala de timp printr-un sir de numere suficient de apropiate, spre exemplu [0 : 0.0005 : 0.03].


```python
# Define the time span and number of samples
time = 0.03  # total time duration
sampling_int = 0.0005  # time between samples
samples = int(time / sampling_int)  # nr of samples
print(samples)
```

```python
# Create time array
t = np.linspace(0, time, samples)  
```

### b) Construiti semnalele x(t), y(t) si z(t) si afisati-le grafic, in cate unsubplot.


```python
# Generate signals

frequency_x = 260 
frequency_y = 140  
frequency_z = 60

phase_x = np.pi / 3
phase_y = -np.pi / 3
phase_z = np.pi / 3

x = np.cos(2 * np.pi * frequency_x * t + phase_x)
y = np.cos(2 * np.pi * frequency_y * t + phase_y)
z = np.cos(2 * np.pi * frequency_z * t + phase_z)
```

```python
# Plot the signals in subplots

figure, axis = plt.subplots(3, sharex=True, figsize=(10, 5))

axis[0].plot(t, x, color='#241468')
axis[0].set_title('Signal x')
axis[1].plot(t, y, color='#9F0D7F')
axis[1].set_title('Signal y')
axis[2].plot(t, z, color='#EA1179')
axis[2].set_title('Signal z')

axis[2].set_xlabel('Time (s)')

for ax in axis:
    ax.grid(True)

plt.tight_layout()
plt.show()
```

### c) Esantionati semnalele cu o frecventa de 200 Hz pentru a obtine x[n], y[n] si z[n] si afisati-le grafic, in cate un subplot.

```python
# Define the time span and number of samples

frequency = 200
print(2 * np.pi / frequency)
samples2 = round(0.03 * frequency) + 1
print(samples2)
```

### Generati urmatoarele semnale si afisati-le grafic, fiecare intr-un plot:

### a) Un semnal sinusoidal de frecventa 400 Hz, care sa contina 1600 de esantioane.

```python
# Define the frequency, number of samples and the time stamp

frequency = 400
samples = 1600
period = 1.0 / frequency
sampling_period = period / samples
```

```python
# Create time array
time = np.linspace(0, samples) * sampling_period
```

```python
# Generate sinusoidal signal
x = np.sin(2 * np.pi * frequency * time)
```

```python
# Plot the signal
plt.figure(figsize=(10, 5))
plt.plot(time, x, color='#9F0D7F')
plt.title('400 Hz Sinusoidal Signal')
plt.xlabel('Time (s)')
plt.grid(True)
plt.show()
```

### b) Un semnal sinusoidal de frecventa 800 Hz, care sa dureze 3 secunde.

```python
# Define the frequency, number of samples and the time stamp
frequency = 800
total_time = 3.0
samples = int(frequency * total_time)
```

```python
# Create time array
time = np.linspace(0, total_time, samples, endpoint=False)
```

```python
# Generate signal
x = np.sin(2 * np.pi * frequency * time)
```

```python
# Plot the signal
plt.figure(figsize=(10, 5))
plt.plot(time, x, color='#241468')
plt.title('800 Hz Sinusoidal Signal for 3s')
plt.xlabel('Time (s)')
plt.grid(True)
plt.show()
```

### c) Un semnal de tip sawtooth de frecventa 240 Hz (puteti folosi functiile numpy.floor sau numpy.mod).

```python
# Define the frequency, number of samples and the time stamp
frequency = 240
total_time = 1.0
samples = int(44100 * total_time)
```

```python
# Create time array
time = np.linspace(0, total_time, samples, endpoint=False)
```

```python
# Generate signal
x = 2 * (time * frequency - np.floor(0.5 + time * frequency))
```

```python
# Plot the signal
plt.figure(figsize=(10, 5))
plt.plot(time, x, color='#EA1179')
plt.title('240 Hz Sawtooth Signal')
plt.xlabel('Time (s)')
plt.grid(True)
plt.xlim(0, 0.03)
plt.show()
```

### d) Un semnal de tip square de frecventa 300 Hz (puteti folosi functia numpy.sign).

```python
# Define the frequency, number of samples and the time stamp
frequency = 300
total_time = 1.0
samples = int(44100 * total_time)
```

```python
# Create time array
time = np.linspace(0, total_time, samples, endpoint=False)
```

```python
# Generate signal
x = np.sign(np.sin(2 * np.pi * frequency * time))
```

```python
# Plot the signal
plt.figure(figsize=(10, 5))
plt.plot(time, x, color='#9F0D7F')
plt.title('300 Hz Square Signal')
plt.xlabel('Time (s)')
plt.grid(True)
plt.xlim(0, 0.03)
plt.show()
```

### f) Un semnal 2D la alegerea voastra. Creati un numpy.array de dimensiune 128x128 si initializati-l folosind o procedura creata de voi. Utilizati, spre exemplu, functiile numpy.zeros() si numpy.ones().

```python
# Chess board 
# Define the dimensions of the 2D signal
x = 128
y = 128
```

```python
# Generate signal
# Initialize a numpy array with zeros (black) for chess
signal = np.zeros((x, y))
```

```python
# Fill the matrix with value 1 (white) where needed for chess
for i in range(x):
    for j in range(y):
        if (i // 16) % 2 == (j // 16) % 2:
            signal[i][j] = 1
```

```python
# Plot the signal

plt.imshow(signal, cmap='gray')
plt.title('Custom 2D Signal (Chess)')
plt.colorbar()
plt.show()
```
