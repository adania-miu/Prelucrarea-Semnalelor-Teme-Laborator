```python
def signal(t):
    frequency = 5
    s = np.sin(2 * np.pi * frequency * t)
    return s

samples = np.arange(-1, 1, 0.001)
x_list = list(map(signal, samples))
```

```python
def f(samples, w, n):
    return samples(n) * math.e ** (2 * np.pi * 1j * w * n)
```

```python
w = [1, 1, 2, 5, 7]

N = [128, 256, 512, 1024, 2048, 4096, 8192]
```

```python
figure, axes = plt.subplots(4, 2, figsize=(10, 15))

axes = axes.flatten()

for ax in axes:
    ax.set_xlim(-1, 1)
    ax.set_ylim(-1, 1)
    
    ax.set_xticks([-1, -0.5, 0, 0.5, 1])
    ax.set_yticks([-1, -0.5, 0, 0.5, 1])
    
    ax.axhline(0, color='black')
    if ax != axes[0]:
        ax.axvline(0, color='black')
    
axes[0].plot(samples, x_list, color="green")

for i, elem in enumerate(N):
    points = np.fft.fft(x_list, elem)
    reals = [e.real for e in points]
    imags = [e.imag for e in points]
    axes[i+1].plot(imags, reals, color="#9F0D7F")
    
plt.show()
```
